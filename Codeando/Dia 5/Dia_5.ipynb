```python
def suma_menores(lista_numeros):
    """
    Esta función suma los números de una lista que son mayores a 0 y menores a 1000.
    Parámetros:
    lista_numeros: una lista de números.
    Retorno:
    La suma de los números de la lista que son mayores a 0 y menores a 1000.
    """
    suma= 0
    for i in lista_numeros:
        if 0 < i < 1000:
            suma += i
        else:
            pass
    return suma

lista_numeros = [1,2,3,1000,-5]
muestra=suma_menores(lista_numeros)
print(muestra)
```

```python
precios_cafe = [("capuchino",45),("expresso", 20),("moka", 10)]


def cafe_mas_caro(precios_cafe):
    precio_mayor= 0
    nombre_cafe = ""

    for cafe,precio in precios_cafe:
        if precio > precio_mayor:
            precio_mayor = precio
            nombre_cafe = cafe
        else:
            pass

    return (precio_mayor,nombre_cafe)


mostrar= cafe_mas_caro(precios_cafe)
print(mostrar)

precio, cafe = cafe_mas_caro(precios_cafe)

print(f"el cafe mas caro es el {cafe} y su precio es {precio}")
```

Práctica sobre Argumentos Indefinidos (*args) 1

Crea una función llamada suma_cuadrados que tome una cantidad indeterminada de argumentos numéricos, y que retorne la suma de sus valores al cuadrado.

Por ejemplo para los argumentos suma_cuadrados(1,2,3) deberá retornar 14 (1+4+9).

```python
def suma_cuadrados(*args):
    lista = []
    cuadradros =0
    for i in args:
        cuadradros = i**2
        lista.append(cuadradros)
    return lista, sum(lista)

print(suma_cuadrados(1,2,3,4))
```

```python
def suma_absolutos(*args):
    suma=0
    for i in args:
        suma += abs(i)
    return suma

print(suma_absolutos(-1,2,3,4,-5))
```

Crea una función llamada lista_atributos que devuelva en forma de lista los valores de los atributos entregados en forma de palabras clave (keywords). La función debe preveer recibir cualquier cantidad de argumentos de este tipo.

```python
def lista_atributos(**kwargs):
    lista = []
    for j in kwargs.values():
        lista.append(j)
    return lista

print(lista_atributos(z="uno",x="dos",a="tres"))

print("************************************")

listaaa={"a":"cinco",
        "b":"dos",
        "c":"tres",
        "d":"cuatro",
        "e":"uno"}

print(lista_atributos(**listaaa))
```

Práctica sobre Argumentos Indefinidos (**kwargs) 3

Crea una función llamada describir_persona, que tome como parámetros su nombre y luego una cantidad indetermida de argumentos. Esta función deberá mostrar en pantalla:

Características de {nombre}:
{nombre_argumento}: {valor_argumento}
{nombre_argumento}: {valor_argumento}
etc...
Por ejemplo:

describir_persona("María", color_ojos="azules", color_pelo="rubio")

Mostrará en pantalla:

Características de María:
color_ojos: azules
color_pelo: rubio

```python
def describir_persona(nombre,**kwargs):
    print(f"Caracteristicas de {nombre}")
    for clave,valor in kwargs.items():
        print(f"su {clave} es {valor}")

nombre= "Daniel"
diccionario={
    "edad":25,
    "altura":1.80,
    "color pelo":"Negro",
    "Color ojos":"Negro"}

print(describir_persona(nombre,**diccionario))
```

```python
def devolver_distintos(num_1,num_2,num_3):
    lista = [num_1,num_2,num_3]
    suma = num_1+ num_2 + num_3
    if suma > 15:
        return max(lista)
    elif suma < 10:
        return min(lista)
    else:
        lista.sort()
        return lista[1]

print(devolver_distintos(2,4,6))

```
