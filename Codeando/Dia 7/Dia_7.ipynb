# Prctica Tipos de Mtodos 3
Crea un mtodo de instancia lanzar_flecha() que reste en -1 la cantidad de flechas que tiene una instancia de Personaje, que cuenta con un atributo de instancia de tipo nmero, llamado cantidad_flechas.

```python
class Jugador:
    
    def __init__(self, cantidad_flechas):
        self.cantidad_fechas = cantidad_flechas
        
    def lanzar_fecha(self):
        if self.cantidad_fechas > 0:
            self.cantidad_fechas -= 1
            print(f"te quedan {self.cantidad_fechas} felchas")
        else:
            print("no hay mas flechas por lanzar")
            
mi_jugador = Jugador(10)

while mi_jugador.cantidad_fechas>=0:
    mi_jugador.lanzar_fecha()
    if mi_jugador.cantidad_fechas == 0:
        break
```

# Prctica Mtodos Especiales 3
Dada la clase Libro, implementa el mtodo especial __del__ para que el usuario sea informado con el mensaje "Libro eliminado", mostrndolo en pantalla cada vez que el libro se elimine.

```python
class Libro():
    def __init__(self, titulo, autor, cantidad_paginas):
        self.titulo = titulo
        self.autor = autor
        self.cantidad_paginas = cantidad_paginas
 
    def __str__(self):
        return f'"{self.titulo}", de {self.autor}'
 
    def __len__(self):
        return self.cantidad_paginas
 
    def __del__(self):
        return print("Libro eliminado")
 
 
libro1 = Libro('La Casa del Dragon', 'George RR Martin', 883)
print(libro1, " Paginas: ", len(libro1))
del libro1
```
