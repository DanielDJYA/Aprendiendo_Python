```python
mi_tuple = (1,2,3,4,5)
print(type(mi_tuple))
print(mi_tuple)
print(mi_tuple[2])

#mi_tuppe[0]=11 # aca saldra error, los tuple una vez creado no puede ser modificado, es aprueba de daños o cambios
#print(mi_tuppe)

mi_tuple_dos=(1,2,3,4,5,(10,11),500)
print(mi_tuple_dos[5])
print(mi_tuple_dos[5][0])
print(mi_tuple_dos[5][1])

#podemos convertir un tuple a una lista y a la inversa

a = list(mi_tuple_dos)
print(a)

b= tuple(a)     # aca cambia en [] la lista va entre []
print(b)        # aca cambia en ()) los tuple va entre  ()

#recordar tambien que se pueden aplicar funciones 
#por ejemplo .count() que permite contar las apariciones de un valor dentro de tu tuppe
print(b.count(1)) 
print(b.count(3))
```
