```python
diccionario = {
    "Nombre":"Daniel",
    "Apellido":"Yañez",
    "Edad":"25"}
for i in diccionario:
    print(i)
print("*****************************")
for i in diccionario.items(): #imprime todo
    print(i)
print("*****************************")
for i in diccionario.keys(): #imprime solo las llaves
    print(i)
print("*****************************")
for i in diccionario.values(): #imprime solo los valores
    print(i)
print("*****************************")
for i,j in diccionario.items(): #imprime todo
    print(f"el valor de {i} es {j}")
print("*****************************")
```
